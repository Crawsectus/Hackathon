```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset_Talento.csv', parse_dates=['timestamp'])
```

```python
df=df.drop(columns={"observaciones","temperatura","vibración","humedad","tiempo_ciclo","fallo_detectado","tipo_fallo","consumo_energia","paradas_programadas","paradas_imprevistas","timestamp"})
```

```python
df['turno'] = df['turno'].map({
    'Mañana': 1,
    'Tarde': 2,
    'Noche': 3
})
df['operador_id'] = df['operador_id'].str.replace(r'\D+', '', regex=True)
df['operador_id'] = df['operador_id'].astype(int)  
df['maquina_id'] = df['maquina_id'].str.replace(r'\D+', '', regex=True)
df['maquina_id'] = df['maquina_id'].astype(int)  
df['producto_id'] = df['producto_id'].str.replace(r'\D+', '', regex=True)
df['producto_id'] = df['producto_id'].astype(int)  
df.head()
```

```python
df['eficiencia_porcentual'] = df['eficiencia_porcentual'].fillna(84.97)
df.to_csv("DatosLimpios.csv",index=False)
```

```python
df.isnull().sum()
```

```python
# Calcular productos buenos
df['producto_bueno'] = df['cantidad_producida'] - df['unidades_defectuosas']

# Listas únicas
operadores = df['operador_id'].unique()
maquinas = df['maquina_id'].unique()
productos = df['producto_id'].unique()
turnos = df['turno'].unique()
turnosdict = {1: "mañana", 2: "tarde", 3: "noche"}

# Lista para guardar resultados
resultados = []

# Recorrer combinaciones y guardar el mejor caso por operador
for operador in sorted(operadores):
    max_prod = 0
    mejor_maquina = 0
    mejor_turno = 0
    mejor_prod = 0
    
    for maquina in sorted(maquinas):
        for producto in sorted(productos):
            for turno in sorted(turnos):
                filtro = (
                    (df['operador_id'] == operador) &
                    (df['maquina_id'] == maquina) &
                    (df['producto_id'] == producto) &
                    (df['turno'] == turno)
                )
                total_buenos = df[filtro]['producto_bueno'].sum()
                
                if total_buenos > max_prod:
                    mejor_turno = turno
                    mejor_prod = producto
                    mejor_maquina = maquina
                    max_prod = total_buenos

    resultados.append({
        'Operador': operador,
        'Mejor Máquina': mejor_maquina,
        'Mejor Turno': turnosdict[mejor_turno],
        'Mejor Producto': mejor_prod,
        'Productos Buenos': max_prod
    })

# Convertir a DataFrame
tabla_resultados = pd.DataFrame(resultados)

# Mostrar tabla ordenada por operador
print(tabla_resultados.sort_values(by="Operador"))
```

```python
from IPython.display import display

display(tabla_resultados.sort_values(by="Operador"))
```

```python
import seaborn as sns

# Ordenar operadores por cantidad de productos buenos
tabla_ordenada = tabla_resultados.sort_values(by='Productos Buenos', ascending=False)

# Crear gráfico
plt.figure(figsize=(12, 6))
sns.barplot(data=tabla_ordenada, x='Operador', y='Productos Buenos', hue='Mejor Máquina', palette='Set2')

# Estética
plt.title('Máxima producción de productos buenos por operador')
plt.xlabel('Operador')
plt.ylabel('Productos buenos (máximo por configuración)')
plt.legend(title='Máquina')
plt.tight_layout()
plt.show()
```
