# Reto Hackathon análisis de datos talento tech

## Importación de librerías

```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

## Lectura y limpieza de datos

```python
df = pd.read_csv('Dataset_Talento.csv', parse_dates=['timestamp'])
```

Eliminamos las columnas con comportamiento demasiado lineal

```python
df=df.drop(columns={"observaciones","temperatura","vibración","humedad","tiempo_ciclo","fallo_detectado","tipo_fallo","consumo_energia","paradas_programadas","paradas_imprevistas","timestamp"})
```

Convertimos los campos de string a tipos numéricos

```python
df['turno'] = df['turno'].map({
    'Mañana': 1,
    'Tarde': 2,
    'Noche': 3
})
df['operador_id'] = df['operador_id'].str.replace(r'\D+', '', regex=True)
df['operador_id'] = df['operador_id'].astype(int)  
df['maquina_id'] = df['maquina_id'].str.replace(r'\D+', '', regex=True)
df['maquina_id'] = df['maquina_id'].astype(int)  
df['producto_id'] = df['producto_id'].str.replace(r'\D+', '', regex=True)
df['producto_id'] = df['producto_id'].astype(int)  
df.head()
```

Eficiencia poseía 180 filas vacías, se rellena estos campos vacíos con el promedio de la eficiencia

```python
df['eficiencia_porcentual'] = df['eficiencia_porcentual'].fillna(84.97)
df.to_csv("DatosLimpios.csv",index=False)
```

Comprobamos que ya no hay elementos nulos

```python
df.isnull().sum()
```

Consultamos cual es el mejor resultado ofrecido por cada empleado

```python
df['producto_bueno'] = df['cantidad_producida'] - df['unidades_defectuosas']

operadores = df['operador_id'].unique()
maquinas = df['maquina_id'].unique()
productos = df['producto_id'].unique()
turnos = df['turno'].unique()
turnosdict = {1: "mañana", 2: "tarde", 3: "noche"}

resultados = []

for operador in sorted(operadores):
    max_prod = 0
    mejor_maquina = 0
    mejor_turno = 0
    mejor_prod = 0
    
    for maquina in sorted(maquinas):
        for producto in sorted(productos):
            for turno in sorted(turnos):
                filtro = (
                    (df['operador_id'] == operador) &
                    (df['maquina_id'] == maquina) &
                    (df['producto_id'] == producto) &
                    (df['turno'] == turno)
                )
                total_buenos = df[filtro]['producto_bueno'].sum()
                
                if total_buenos > max_prod:
                    mejor_turno = turno
                    mejor_prod = producto
                    mejor_maquina = maquina
                    max_prod = total_buenos

    resultados.append({
        'Operador': operador,
        'Mejor Máquina': mejor_maquina,
        'Mejor Turno': turnosdict[mejor_turno],
        'Mejor Producto': mejor_prod,
        'Productos Buenos': max_prod
    })

tabla_resultados = pd.DataFrame(resultados)

print(tabla_resultados.sort_values(by="Operador"))
```

```python
from IPython.display import display

display(tabla_resultados.sort_values(by="Operador"))
```

```python
import seaborn as sns

# Ordenar operadores por cantidad de productos buenos
tabla_ordenada = tabla_resultados.sort_values(by='Productos Buenos', ascending=False)

# Crear gráfico
plt.figure(figsize=(12, 6))
sns.barplot(data=tabla_ordenada, x='Operador', y='Productos Buenos', hue='Mejor Máquina', palette='Set2')

# Estética
plt.title('Máxima producción de productos buenos por operador')
plt.xlabel('Operador')
plt.ylabel('Productos buenos (máximo por configuración)')
plt.legend(title='Máquina')
plt.tight_layout()
plt.show()
```

Realizamos una simulación de cuanto produciría cada empleado si se mantuviera en su mejor turno, máquina y producto, y lo comparamos con los resultados reales

```python
predicciones = []

for operador in sorted(operadores):
    registros_operador = df[df['operador_id'] == operador]
    total_turnos = len(registros_operador)

    fila = tabla_resultados[tabla_resultados['Operador'] == operador].iloc[0]
    mejor_maquina = fila['Mejor Máquina']
    mejor_producto = fila['Mejor Producto']
    mejor_turno = [k for k, v in turnosdict.items() if v == fila['Mejor Turno']][0]

    filtro_mejor = (
        (df['operador_id'] == operador) &
        (df['maquina_id'] == mejor_maquina) &
        (df['producto_id'] == mejor_producto) &
        (df['turno'] == mejor_turno)
    )
    registros_mejor_config = df[filtro_mejor]

    if not registros_mejor_config.empty:
        promedio_buenos = registros_mejor_config['producto_bueno'].mean()
        prediccion_total = promedio_buenos * total_turnos

        predicciones.append({
            'Operador': operador,
            'Mejor Máquina': mejor_maquina,
            'Mejor Turno': fila['Mejor Turno'],
            'Mejor Producto': mejor_producto,
            'Promedio por turno (en mejor config)': round(promedio_buenos, 2),
            'Turnos totales': total_turnos,
            'Predicción total (sin rotar)': int(prediccion_total)
        })

tabla_prediccion = pd.DataFrame(predicciones)
reales = df.groupby('operador_id')['producto_bueno'].sum().reset_index()
reales.columns = ['Operador', 'Total Real']
comparacion = pd.merge(tabla_prediccion, reales, on='Operador')

comparacion = comparacion[['Operador', 'Total Real', 'Predicción total (sin rotar)']]
import matplotlib.pyplot as plt
import numpy as np

operadores = comparacion['Operador'].astype(str)
reales = comparacion['Total Real']
predicciones = comparacion['Predicción total (sin rotar)']

x = np.arange(len(operadores))
width = 0.35

plt.figure(figsize=(14, 6))
plt.bar(x - width/2, reales, width, label='Producción real', color='skyblue')
plt.bar(x + width/2, predicciones, width, label='Predicción sin rotar', color='orange')

plt.xlabel('Operador')
plt.ylabel('Productos buenos')
plt.title('Comparación: Producción real vs Predicción sin rotación por operador')
plt.xticks(x, operadores)
plt.legend()
plt.tight_layout()
plt.show()
```
